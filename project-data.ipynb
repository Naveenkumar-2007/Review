```python
df=data[['reviewText','rating']]
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
df['rating']=df['rating'].apply(lambda x: 0 if x<3 else 1)
```

```python
df['rating'].unique()
```

```python
df.head(28)
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
##to remove special cherectors
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+','',x))
##to remove stopwords
df['reviewText']=df['reviewText'].apply(lambda x: " ".join([y for y in x.split() if y not in stopwords.words('english')]))
```

```python
from bs4 import BeautifulSoup
```

```python
#to remove the links
df['reviewText']=df['reviewText'].apply(lambda x:re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?','',str(x)))
#to remove html tages 
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
## Remove any additional spaces
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
from nltk.stem import WordNetLemmatizer
word=WordNetLemmatizer()
```

```python
def word_lem(text):
    return " ".join(word.lemmatize(wor) for wor in text.split())
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:word_lem(x))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
con=CountVectorizer()
x_train_bow=con.fit_transform(x_train)
x_test_bow=con.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mul=MultinomialNB()
mul.fit(x_train_bow,y_train)
```

```python
ypre=mul.predict(x_test_bow)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,ypre))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
x_train_tf=tf.fit_transform(x_train)
x_test_tf=tf.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
muli=MultinomialNB()
muli.fit(x_train_tf,y_train)
```

```python
ypree=muli.predict(x_test_tf)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,ypree))
```

```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# STEP 3: Train Word2Vec
w2v_model = Word2Vec(sentences=x_train, vector_size=100, window=5, min_count=2, sg=1)

# STEP 4: Convert sentence to vector using average word embeddings
def get_avg_vector(tokens, model, vector_size=100):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(vector_size)
    return np.mean(vectors, axis=0)

# Convert all train/test data to vectors
x_train_vec = np.array([get_avg_vector(tokens, w2v_model) for tokens in x_train])
x_test_vec = np.array([get_avg_vector(tokens, w2v_model) for tokens in x_test])

# STEP 5: Train ML model (Logistic Regression)
clf = GaussianNB()
clf.fit(x_train_vec, y_train)

# STEP 6: Predict and show accuracy
y_pred = clf.predict(x_test_vec)
print(" Accuracy:", accuracy_score(y_test, y_pred))
```
